```python
import cv2
import numpy as np
from mss import mss

# Define the screen region where obstacles appear (adjust to your game)
obstacle_region = {'top':205, 'left':647, 'width':243, 'height':175}

# Initialize screen capture
cap = mss()

def obstacle_present():
    region_img = np.array(cap.grab(obstacle_region))[:, :, :3]
    gray = cv2.cvtColor(region_img, cv2.COLOR_BGR2GRAY)
    _, thresh = cv2.threshold(gray, 170, 255, cv2.THRESH_BINARY)
    resized = cv2.resize(thresh, (100, 83))
    obstacle_pixels = np.sum(resized == 255)

    # Show the thresholded image for tuning
    cv2.imshow("Obstacle Threshold", thresh)

    # Print pixel count
    # print(f"White pixels (obstacle): {obstacle_pixels}")

    # Return detection status
    return obstacle_pixels > 25  # adjust based on what you observe

def run_detection():
    print("Press 'q' to quit.")
    while True:
        is_obstacle = obstacle_present()
        if is_obstacle:
            print("Obstacle Detected!" )
        else:
            print("no")

        # Break on 'q'
        if cv2.waitKey(30) & 0xFF == ord('q'):
            break

    cv2.destroyAllWindows()

if __name__ == "__main__":
    run_detection()
```
